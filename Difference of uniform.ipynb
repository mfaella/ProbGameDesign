## Computing the distribution of the difference of two discrete uniform variables

```python
import matplotlib.pyplot as plt

# Plots the distribution of the difference of two discrete
# uniform RVs on the range 0...max
max = 255;
freq = {}

for i in range(0, max+1):
    for j in range(0, max+1):
        diff = i * j;
        freq[diff] = freq.get(diff, 0)+1;

normalized_freq = [x / 2 for x in freq.values()]        
        
# The plot
fig, ax = plt.subplots()
ax.plot(list(freq.keys()), list(freq.values()))
ax.set(xlabel='n', ylabel='Pr(n)', title='Distribution of the difference of two uniform')
ax.grid()
plt.show()
```
